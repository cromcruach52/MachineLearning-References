```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Sample data (size of houses in square feet

house_sizes = np.array([1000, 1500, 2000, 2500, 3000]).reshape(-1, 1)

# Corresponding sale prices (in dollars

sale_prices = np.array([200000, 300000, 400000, 500000, 600000])
```

```python
# Plot Data (figsize = size of the figure, scaling of the image per say)

plt.figure(figsize=(8,6))

#plotting scatter to test (You can check label to see what other functions we can use

plt.scatter(house_sizes, sale_prices, color = 'red', label = 'Data Points')
plt.title('House Size V.S. Sale Price')

# x label and y label are x and y coordinate name

plt.xlabel('Size')
plt.ylabel('Sale Price')

# plt.legend is an indicator for the things we used, i.e., label showed up - Data Point

plt.legend()
```

```python
# fit linear regression model

model = LinearRegression()
model.fit(house_sizes, sale_prices)

# plot the linear regression model

plt.plot(house_sizes, model.predict(house_sizes), color = 'blue', label = 'linear regression')
plt.legend()

plt.show()
```

# Histogram

```python
# Using imports from URL

import pandas as pd
```

```python
# Load the dataset

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"


# Column names are not included in the dataset, so we add them manually

column_names = ['Class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium',
                'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins',
                'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline']

data = pd.read_csv(url, names=column_names)
```

```python
# column of interest, selected columns that the user specified

column_of_interest = 'Alcohol'


plt.figure(figsize = (10,6))
plt.hist(data[column_of_interest], bins = 20, color = 'cyan', edgecolor = 'black', alpha = 0.7)
plt.title('Histogram of ' + column_of_interest)

# x label and y label are x and y coordinate names

plt.xlabel(column_of_interest)
plt.ylabel('Frequency')
plt.show()
```

# Line Chart

```python
# Example: Identifying a trend with a line chart
# random normal - different from everyone else since its normal

x = np.linspace(0, 10, 100)
y = 2 * x + 3 + np.random.normal(0, 2, 100)

plt.figure(figsize = (8,6))
plt.plot(x, y, label = 'Data')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Identifying a Trend with Data Visualization')
plt.legend()

plt.show()
```

# Scatter Plot


```python
# import data from URL

iris_data = pd.read_csv("https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv")
```

```python
iris_data.head()
```

```python
# Plotting Iris Data with Scatter Plot

plt.figure(figsize = (10,6))
plt.scatter(iris_data['sepal_length'], iris_data['sepal_width'], c = iris_data['species'].astype('category').cat.codes, cmap = 'viridis')

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal length V.S. Sepal Width')
plt.show()
```

# Pie Chart

```python
# Load dataset

wine_data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data",
                        names=['Class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash',
                               'Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols',
                               'Proanthocyanins', 'Color intensity', 'Hue',
                               'OD280/OD315 of diluted wines', 'Proline'])
```

```python
wine_data.head()
```

```python
plt.figure(figsize = (8,6))
class_counts = wine_data['Class'].value_counts()
plt.pie(class_counts, labels = class_counts.index, autopct = '%1.1f%%', startangle = 140)
plt.title('Distribution of Wine Classes')
plt.axis('equal')
plt.show()
```
