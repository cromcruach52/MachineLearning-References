# Original Image

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Read the noisy image or path
noisy_image = cv2.imread('Images/noisy2.jpg')

plt.imshow(cv2.cvtColor(noisy_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

# Gaussian filter applied

```python
gaussian_filter_img = cv2.GaussianBlur(noisy_image, (5,5), 0)


plt.imshow(cv2.cvtColor(gaussian_filter_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

# Contrast stretching applied

```python
image =  cv2.imread('Images/xray.png', cv2.IMREAD_GRAYSCALE)
min_val = np.min(image)
max_val = np.max(image)

# show image for grayscale
cv2.imshow("Grayscale", image)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

# Contrast stretching


```python
contrast_stretched_image = np.uint8(255 * ((image- min_val) / (max_val - min_val)))

cv2.imshow("Contrast Stretched Image", contrast_stretched_image)
plt.imshow(cv2.cvtColor(contrast_stretched_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

# Normalization

```python
normalize_image = cv2.normalize(image, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)

cv2.imshow("Normalized Image", normalize_image)
plt.imshow(cv2.cvtColor(normalize_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

# Resizing

```python
# Resize the image
resized_image = cv2.resize(image, (266,266))


# Display the resized image
cv2.imshow("Resized Image", resized_image)
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

```

# Grayscale

```python
# Convert the image to grayscale
image2 =  cv2.imread('Images/xray.png')
gray_image = cv2.cvtColor(image2,cv2.COLOR_BGR2GRAY)

# Apply segmentation
_, segmented_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

# Display the segmented image
cv2.imshow("Segmented Image", segmented_image)
plt.imshow(cv2.cvtColor(segmented_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

```

# Edges


```python
gray_image = cv2.cvtColor(image2,cv2.COLOR_BGR2GRAY)

# Apply feature extraction
edges = cv2.Canny(gray_image, threshold1=100,  threshold2=200)

# Display the edges
cv2.imshow("Edges", edges)
plt.imshow(cv2.cvtColor(edges, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

# Data Augmentation

```python
# Perform data augmentation

flipped_image = cv2.flip(image, 1)

# Display the edges
cv2.imshow("Original Image", image)
cv2.imshow("Flipped Image", flipped_image)
plt.imshow(cv2.cvtColor(flipped_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

```
